```python
import pandas as pd
from sklearn.preprocessing import OrdinalEncoder, LabelEncoder
```

```python

columns = ['buying', 'maint', 'doors', 'persons', 'lug_boot', 'safety', 'class']

df = pd.read_csv("car_evaluation.csv", header=None, names=columns)
df.head()
```

```python
# Ordinal categories
buying_order = ['low', 'med', 'high', 'vhigh']
safety_order = ['low', 'med', 'high']
```

```python
ord_enc = OrdinalEncoder(categories=[buying_order, safety_order])
```

```python
df[['buying_enc', 'safety_enc']] = ord_enc.fit_transform(
    df[['buying', 'safety']]
)
```

```python
le = LabelEncoder()

for col in ['maint', 'doors', 'persons', 'lug_boot', 'class']:
    df[col + '_enc'] = le.fit_transform(df[col])
```
